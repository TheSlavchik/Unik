```python
import numpy as np

# Исходные данные
data = [
    41.7, 36.7, 45.6, 31.0, 38.9, 28.3, 23.7, 54.5, 31.6, 26.1,
    50.4, 45.0, 53.5, 50.1, 29.7, 40.9, 32.4, 45.0, 35.6, 25.6,
    55.3, 33.1, 25.6, 40.2, 32.2, 43.2, 25.1, 29.9, 52.7, 18.6,
    35.3, 38.1, 29.4, 32.1, 43.0, 23.0, 34.1, 28.8, 31.6, 40.8,
    32.7, 43.0, 41.7, 41.7, 28.3, 23.1, 18.2, 23.5, 30.3, 36.2,
    34.9, 41.0, 54.3, 33.3, 39.3, 12.8, 20.6, 22.6, 26.1, 16.8
]

# Оценка среднего (mu)
mu_hat = np.mean(data)

# Оценка дисперсии (sigma^2) методом правдоподобия (смещенная)
# В методе макс. правдоподобия делится на n, а не на n-1
sigma_sq_hat = np.var(data) 

# Стандартное отклонение (sigma)
sigma_hat = np.sqrt(sigma_sq_hat)

print(f"Оценка математического ожидания (mu): {mu_hat:.4f}")
print(f"Оценка дисперсии (sigma^2): {sigma_sq_hat:.4f}")
print(f"Оценка стандартного отклонения (sigma): {sigma_hat:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm, chi2
import math

# Устанавливаем стиль графиков
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.size'] = 12

# Исходные данные (внимание: в исходной строке есть точка вместо запятой в "31.6", исправляем на запятую)
data_str = """41,7; 36,7; 45,6; 31,0; 38,9; 28,3; 23,7; 54,5; 31,6; 26,1; 50,4; 45,0; 53,5; 50,1; 29,7; 40,9; 32,4; 45,0; 35,6; 25,6; 55,3; 33,1; 25,6; 40,2; 32,2; 43,2; 25,1; 29,9; 52,7; 18,6; 35,3; 38,1; 29,4; 32,1; 43,0; 23,0; 34,1; 28,8; 31,6; 40,8; 32,7; 43,0; 41,7; 41,7; 28,3; 23,1; 18,2; 23,5; 30,3; 36,2; 34,9; 41,0; 54,3; 33,3; 39,3; 12,8; 20,6; 22,6; 26,1; 16,8"""

# Парсим данные
data = np.array([float(x.replace(',', '.')) for x in data_str.split(';')])
n = len(data)
print(f"Объем выборки n = {n}")
print(max(data))
print(min(data))

# 1. Вариационный ряд
var_series = np.sort(data)
print("\n1. Вариационный ряд:")
print(var_series)

# 2. Эмпирическая функция распределения
x_ecdf = np.sort(data)
y_ecdf = np.arange(1, n+1) / n

# Строим график ЭФР в отдельном окне
plt.figure(figsize=(10, 6))
plt.step(x_ecdf, y_ecdf, where='post', label='Эмпирическая F*(x)', color='blue', linewidth=2)
plt.title('2. Эмпирическая функция распределения', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('F*(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# 3. Числовые характеристики
mean_val = np.mean(data)
var_val = np.var(data, ddof=1)  # исправленная (несмещенная) дисперсия
std_val = np.std(data, ddof=1)
# Мода (приближенная для непрерывных данных - округляем до 1 знака после запятой)
rounded_data = np.round(data, 1)
mode_result = stats.mode(rounded_data, keepdims=False)
mode_val = mode_result.mode if hasattr(mode_result, 'mode') else mode_result[0]
# Медиана
median_val = np.median(data)
# Эксцесс и асимметрия
skew_val = stats.skew(data)
kurt_val = stats.kurtosis(data, fisher=True) # Fisher = True дает эксцесс - 3 (для нормального 0)

print("\n3. Числовые характеристики:")
print(f"Выборочное среднее: {mean_val:.4f}")
print(f"Исправленная дисперсия: {var_val:.4f}")
print(f"Исправленное СКО: {std_val:.4f}")
print(f"Мода: {mode_val}")
print(f"Медиана: {median_val:.4f}")
print(f"Асимметрия: {skew_val:.4f}")
print(f"Эксцесс: {kurt_val:.4f}")

# 4. Интервальный статистический ряд
# Используем правило Стёрджеса для определения числа интервалов
k = int(math.ceil(1 + 3.322 * np.log10(n)))
print(k)
# Границы интервалов
bins = np.linspace(data.min(), data.max(), k+1)

# 5. Гистограмма и полигон (по сгруппированным данным)
counts, bin_edges = np.histogram(data, bins=bins)
percentages = (counts / counts.sum())
# Для полигона частот берем середины интервалов
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
# Относительные частоты
rel_freq = counts / n
# Плотности относительных частот для гистограммы (высота столбца)
h = bin_edges[1] - bin_edges[0]
hist_heights = rel_freq / h
ninh = (counts / counts.sum() * h)

print("\n4. Интервальный ряд:")
print(f"Число интервалов: {k}")
print("Границы интервалов:", np.round(bin_edges, 2))
print("Частоты:", counts)
print("Проценты:", percentages)
print("NiNH:", ninh)
print(sum(percentages))

# 6. Гипотеза о распределении (Нормальное)
print("\n6. Гипотеза: Нормальное распределение N(a, sigma)")

# 7. Оценки параметров нормального распределения
mu_hat = mean_val
sigma_hat = np.std(data, ddof=1) # используем исправленное СКО
print(f"7. Оценки параметров: mu = {mu_hat:.4f}, sigma = {sigma_hat:.4f}")

# 8. Построение теоретических кривых
x_theor = np.linspace(data.min() - 2, data.max() + 2, 500)
pdf_theor = norm.pdf(x_theor, mu_hat, sigma_hat)
cdf_theor = norm.cdf(x_theor, mu_hat, sigma_hat)

# График 1: Гистограмма и полигон частот
plt.figure(figsize=(10, 6))
plt.bar(bin_edges[:-1], hist_heights, width=h, align='edge', alpha=0.6, 
        color='skyblue', edgecolor='black', label='Гистограмма')
plt.plot(bin_centers, hist_heights, 'o-', color='red', linewidth=2, 
         markersize=8, label='Полигон частот')
plt.title('5. Гистограмма и полигон относительных частот', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
#plt.ylabel('Плотность относительной частоты', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# График 2: Эмпирическая функция с теоретической
plt.figure(figsize=(10, 6))
plt.step(x_ecdf, y_ecdf, where='post', label='Эмпирическая F*(x)', color='blue', linewidth=2)
plt.plot(x_theor, cdf_theor, 'r--', linewidth=2, label=f'Теоретическая F(x) N({mu_hat:.2f}, {sigma_hat:.2f})')
plt.title('8. Эмпирическая и теоретическая функции распределения', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('F(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# График 3: Гистограмма с теоретической плотностью
plt.figure(figsize=(10, 6))
plt.hist(data, bins=bins, density=True, alpha=0.6, color='skyblue', 
         edgecolor='black', label='Гистограмма')
plt.plot(x_theor, pdf_theor, 'g-', linewidth=2, 
         label=f'Теоретическая f(x) N({mu_hat:.2f}, {sigma_hat:.2f})')
plt.title('8. Гистограмма и теоретическая плотность распределения', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# 9. Правило "трех сигм"
lower_3s = mu_hat - 3*sigma_hat
upper_3s = mu_hat + 3*sigma_hat
count_in_3s = np.sum((data >= lower_3s) & (data <= upper_3s))
percent_3s = count_in_3s / n * 100
print(f"\n9. Правило 'трех сигм':")
print(f"Интервал [{lower_3s:.2f}, {upper_3s:.2f}]")
print(f"Данных в интервале: {count_in_3s} из {n} ({percent_3s:.1f}%)")
print("Выполняется." if percent_3s > 95 else "Не выполняется строго, но близко.")

# 10. Критерий согласия Пирсона (Хи-квадрат)
print("\n10. Критерий Пирсона:")
# Для корректного применения критерия объединяем крайние интервалы, где ожидаемая частота < 5
# Сначала считаем ожидаемые частоты для исходных интервалов
expected_counts = np.zeros(k)
for i in range(k):
    p_low = norm.cdf(bin_edges[i], mu_hat, sigma_hat)
    p_high = norm.cdf(bin_edges[i+1], mu_hat, sigma_hat)
    expected_counts[i] = n * (p_high - p_low)

# Объединяем интервалы
merged_counts = []
merged_expected = []
merged_bins = []

i = 0
while i < len(counts):
    c_sum = counts[i]
    e_sum = expected_counts[i]
    bin_start = bin_edges[i]
    j = i + 1
    while j < len(counts) and e_sum < 5:
        c_sum += counts[j]
        e_sum += expected_counts[j]
        j += 1
    # если последний интервал все еще < 5, объединяем с предыдущим
    if j == len(counts) and e_sum < 5 and len(merged_counts) > 0:
        merged_counts[-1] += c_sum
        merged_expected[-1] += e_sum
        merged_bins[-1] = (merged_bins[-1][0], bin_edges[j])
    else:
        merged_counts.append(c_sum)
        merged_expected.append(e_sum)
        merged_bins.append((bin_start, bin_edges[j]))
    i = j

# Вычисляем статистику Хи-квадрат
chi2_stat = 0
for obs, exp in zip(merged_counts, merged_expected):
    if exp > 0:
        chi2_stat += (obs - exp)**2 / exp

df = len(merged_counts) - 2 - 1  # r - s - 1 = кол-во интервалов - число оцененных параметров (2) - 1
chi2_crit = stats.chi2.ppf(0.95, df)
p_value = 1 - stats.chi2.cdf(chi2_stat, df)

print(f"Объединенные интервалы: {len(merged_counts)}")
print(f"Наблюдаемые частоты: {merged_counts}")
print(f"Ожидаемые частоты: {np.round(merged_expected, 2)}")
print(f"Статистика Хи-квадрат: {chi2_stat:.4f}")
print(f"Критическое значение (a=0.05, df={df}): {chi2_crit:.4f}")
print(f"P-value: {p_value:.4f}")

if chi2_stat < chi2_crit:
    print("Гипотеза о нормальном распределении НЕ ОТВЕРГАЕТСЯ.")
else:
    print("Гипотеза о нормальном распределении ОТВЕРГАЕТСЯ.")

# 11. Доверительные интервалы (гамма = 0.95)
gamma = 0.95
alpha = 1 - gamma
t_crit = stats.t.ppf(1 - alpha/2, n-1)
chi2_low = stats.chi2.ppf(alpha/2, n-1)
chi2_high = stats.chi2.ppf(1 - alpha/2, n-1)

# Для среднего
delta_mean = t_crit * (sigma_hat / np.sqrt(n))
ci_mean = (mean_val - delta_mean, mean_val + delta_mean)

# Для СКО
ci_std_low = np.sqrt((n-1) * var_val / chi2_high)
ci_std_high = np.sqrt((n-1) * var_val / chi2_low)

print("\n11. Доверительные интервалы (уровень доверия 0.95):")
print(f"Для генеральной средней: [{ci_mean[0]:.4f}, {ci_mean[1]:.4f}]")
print(f"Для генерального СКО: [{ci_std_low:.4f}, {ci_std_high:.4f}]")

# График 4: QQ-Plot для проверки нормальности
plt.figure(figsize=(10, 8))
stats.probplot(data, dist="norm", plot=plt)
plt.title('QQ-Plot для проверки нормальности распределения', fontsize=14, fontweight='bold')
plt.xlabel('Теоретические квантили', fontsize=12)
plt.ylabel('Выборочные квантили', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# График 5: Гистограмма с доверительными интервалами
plt.figure(figsize=(10, 6))
plt.hist(data, bins=bins, density=True, alpha=0.6, color='skyblue', 
         edgecolor='black', label='Гистограмма')
plt.plot(x_theor, pdf_theor, 'r-', linewidth=2, label='Теоретическая плотность')
plt.axvline(ci_mean[0], color='green', linestyle='--', linewidth=2, 
            label=f'95% ДИ для среднего: [{ci_mean[0]:.2f}, {ci_mean[1]:.2f}]')
plt.axvline(ci_mean[1], color='green', linestyle='--', linewidth=2)
plt.axvline(mean_val, color='darkgreen', linestyle='-', linewidth=1.5, 
            label=f'Выборочное среднее: {mean_val:.2f}', alpha=0.7)
plt.title('11. Гистограмма с доверительным интервалом для среднего', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()

print("\nВсе графики построены и сохранены в отдельных окнах.")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

def generate_stats_table(data):
    # 1. Вариационный ряд
    x_i = np.sort(data)
    n = len(x_i)
    
    # 2. Эмпирическая функция F_n(x_i)
    f_n = np.arange(1, n + 1) / n
    
    # 3. Теоретическая функция F(x_i) по вашей формуле
    # mu = 34.6467, sigma = 10.104
    # norm.cdf вычисляет Ф((x-mu)/sigma)
    f_theoretical = norm.cdf(x_i, loc=34.6467, scale=10.104)
    
    # 4. Модуль разности
    diff = np.abs(f_n - f_theoretical)
    
    # Сборка таблицы
    df = pd.DataFrame({
        'x_i': x_i,
        'F_n(x_i)': np.round(f_n, 4),
        'F(x_i)': np.round(f_theoretical, 4),
        '|F_n(x) - F(x)|': np.round(diff, 4)
    })
    
    return df

data = [
    41.7, 36.7, 45.6, 31.0, 38.9, 28.3, 23.7, 54.5, 31.6, 26.1, 
    50.4, 45.0, 53.5, 50.1, 29.7, 40.9, 32.4, 45.0, 35.6, 25.6, 
    55.3, 33.1, 25.6, 40.2, 32.2, 43.2, 25.1, 29.9, 52.7, 18.6, 
    35.3, 38.1, 29.4, 32.1, 43.0, 23.0, 34.1, 28.8, 31.6, 40.8, 
    32.7, 43.0, 41.7, 41.7, 28.3, 23.1, 18.2, 23.5, 30.3, 36.2, 
    34.9, 41.0, 54.3, 33.3, 39.3, 12.8, 20.6, 22.6, 26.1, 16.8
]
table = generate_stats_table(data)

print(max(table["|F_n(x) - F(x)|"]))
print(table.to_string(index=False))

```
