```python
def modular_inverse(a, m):
    if m == 1:
        return None
    
    initial_m = m
    x0, x1 = 0, 1
    
    print(f"{'Шаг':<5} | {'Остаток':<10} | {'Частное':<10} | {'Коэффициент x'}")
    print("-" * 50)
    
    step = 1
    while a > 1:
        if m == 0:
            print("\nОшибка: Обратного элемента не существует (числа не взаимно простые).")
            return None
        
        q = a // m
        old_m = m
        
        # Алгоритм Евклида
        a, m = old_m, a % m
        
        # Обновление коэффициента x
        old_x1 = x1
        x1 = x0 - q * x1
        x0 = old_x1
        
        print(f"{step:<5} | {a:<10} | {q:<10} | {x1}")
        step += 1

    # Если результат отрицательный, приводим к положительному по модулю
    if x0 < 0:
        x0 += initial_m
        
    return x0
```

```python
# Пример использования
num = 39283
mod = 131

print(f"Поиск обратного элемента для {num} по модулю {mod}:\n")
result = modular_inverse(num, mod)

if result:
    print(f"\nИтог: Обратный элемент равен {result}")
    print(f"Проверка: ({num} * {result}) % {mod} = {(num * result) % mod}")
```
